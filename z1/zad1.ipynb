```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signal vector x
x = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0, 0])

# Define the length N
N = len(x)

# Construct the IDFT matrix W
W = np.zeros((N, N), dtype=complex)
for m in range(N):
    for k in range(N):
        W[m, k] = np.exp(-1j * 2 * np.pi * m * k / N)

# Calculate the time-domain signal x_mu
x_mu = (1/N) * np.matmul(W, x)

# Plot the synthesized signal
plt.stem(np.arange(N), np.real(x_mu))
plt.title('Synthesized Signal')
plt.xlabel('Time Index')
plt.ylabel('Amplitude')
plt.show()
```
