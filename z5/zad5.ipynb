```python
import numpy as np
import matplotlib.pyplot as plt

# Given information
t = np.linspace(-1, 1, 21)
x = t**2  # c = t^2

# Quantizer function
def quantize(x, Q):
    levels = np.linspace(-1, 1, Q)
    quantized = np.interp(x, levels, levels)
    return quantized

# Plot the original signal and the quantized signals for Q = 9 and Q = 10
plt.figure(figsize=(12, 6))
plt.plot(t, x, label='Original Signal')
plt.plot(t, quantize(x, 9), label='Quantized, Q = 9')
plt.plot(t, quantize(x, 10), label='Quantized, Q = 10')
plt.xlabel('t')
plt.ylabel('x(t)')
plt.title('Original and Quantized Signals')
plt.legend()
plt.grid()
plt.show()

# Calculate the error signal e(k)
e_9 = x - quantize(x, 9)
e_10 = x - quantize(x, 10)

# Plot the error signals and their auto-correlation functions
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(t, e_9)
plt.xlabel('t')
plt.ylabel('e(k), Q = 9')
plt.title('Error Signal')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(t, e_10)
plt.xlabel('t')
plt.ylabel('e(k), Q = 10')
plt.grid()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.acorr(e_9, usevlines=True, maxlags=10, normed=True, lw=2)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation, Q = 9')
plt.title('Autocorrelation of Error Signal')
plt.grid()

plt.subplot(2, 1, 2)
plt.acorr(e_10, usevlines=True, maxlags=10, normed=True, lw=2)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation, Q = 10')
plt.grid()
plt.tight_layout()
plt.show()

# Check the quantizer curve for Q = 9 and Q = 10
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(x, quantize(x, 9))
plt.xlabel('x(t)')
plt.ylabel('Quantized x(t), Q = 9')
plt.title('Quantizer Curve')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, quantize(x, 10))
plt.xlabel('x(t)')
plt.ylabel('Quantized x(t), Q = 10')
plt.grid()
plt.tight_layout()
plt.show()
```
