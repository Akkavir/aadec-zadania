```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

def noise_energy(gamma):
    integrand = lambda t: (t**2)**2
    integral, _ = quad(integrand, -1, 1)
    return gamma**2 * integral

def snr(gamma):
    signal_energy = 2
    noise_energy_value = noise_energy(gamma)
    return 10 * np.log10(signal_energy / noise_energy_value)

gamma_values = np.linspace(0.01, 1, 100)
snr_values = [snr(gamma) for gamma in gamma_values]

plt.plot(gamma_values, snr_values)
plt.xlabel('γ')
plt.ylabel('SNR (dB)')
plt.title('SNR vs γ')
plt.grid()
plt.show()
```
